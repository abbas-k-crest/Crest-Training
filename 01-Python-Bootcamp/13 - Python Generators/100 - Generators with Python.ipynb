```python
def demand_range():
    result = []
    clean = False
    for i in range(1,1001):
        if clean:
            clean = False
            result = []
        if i % 100 == 0:
            result.append(i)
            clean = True
            yield result
        else:
            result.append(i)
```

```python
for _ in demand_range():
    print(_)
```

# Fibonacci 

```python
def fibonacci(n):
    a = 1
    b = 1

    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
for number in fibonacci(10):
    print(number)
```

```python
def simple_gen():
    for x in range(3):
        yield x
```

```python
for n in simple_gen():
    print(n)
```

```python
g = simple_gen()
```
