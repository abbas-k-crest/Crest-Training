```python
def hello():

    print("Inside Hello Function")

    def greet():
        return "Greet function inside hello function"

    def welcome():
        return "Welcome Function inside hello function"

    print(greet())
    print(welcome())

    return welcome
```

```python
hello()
```

```python
welcome()
```

```python
wel = hello()
```

```python
wel
```

```python
def new_decorator(original_func):

    def wrap_func():

        print("Before running the function")
        
        original_func()

        print("After running function")

    return wrap_func
```

```python
def func_need_decorator():
    print("I need a Decorator")
```

```python
decorator_func = new_decorator(func_need_decorator)
```

```python
decorator_func()
```

```python
@new_decorator
def func_need_decorator():
    print("I need a Decorator")
```

```python
func_need_decorator()
```
