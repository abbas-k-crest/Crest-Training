```python
import requests
```

```python
result = requests.get('https://pypi.org/project/beautifulsoup4/')
```

```python
text = result.text
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(text, 'lxml')
```

```python
soup
```

```python
project_description = soup.select('.project-description')
```
