### String Formating

#### .format method

```python
print("Hello my name is {}".format('Abbas'))
```

```python
print("My name is {} and I'm working at {}".format('Abbas', 'Crest Infosystems'))
```

```python
print("My name is {1} and I'm working at {1}".format('Abbas', 'Crest Infosystems'))
```

```python
print("My name is {name} and I'm working at {company}".format(name='Abbas', company='Crest Infosystems'))
```

#### with floating numbers

```python
x = 156/345
```

```python
x
```

```python
print("Calculation value {}".format(x))
```

```python
print("Calculation value {x:1.2f}".format(x=x))
```

```python
print("Calculation value {x:10.2f}".format(x=x))
```

```python
print("Calculation value {x:.4f}".format(x=x))
```
