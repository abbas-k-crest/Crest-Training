```python
class Book:

    def __init__(self, title, author, pages):

        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title} by {self.author}."

    def __len__(self):
        return self.pages

    def __del__(self):
        print(f"{self.title} book has been deleted")
```

```python
mybook = Book("Holy Rocks", "Henry Davidson", 498)
```

```python
str(mybook)
```

```python
len(mybook)
```

```python
print(mybook)
```

```python
del mybook
```
