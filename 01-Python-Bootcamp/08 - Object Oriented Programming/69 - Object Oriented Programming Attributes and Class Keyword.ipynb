```python
class Dog:

    def __init__(self, breed, name, spots):
        self.breed = breed
        self.name = name
        self.spots = spots
```

```python
dog = Dog('Lab', 'Sammy', True)
```

```python
print(dog.breed)
print(dog.spots)
print(dog.name)
```
