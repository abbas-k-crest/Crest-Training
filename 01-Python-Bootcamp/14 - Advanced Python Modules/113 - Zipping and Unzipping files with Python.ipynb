# Zip Individual file

```python
with open('practice_file1.txt', 'w')as f:
    f.write("Hello this is file 1")
```

```python
with open('practice_file2.txt', 'w')as f:
    f.write("Hello this is file 2")
```

```python
import zipfile
```

```python
comp_file = zipfile.ZipFile('comp_file.zip', 'w')
```

```python
comp_file.write('practice_file1.txt', compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.write('practice_file2.txt', compress_type=zipfile.ZIP_DEFLATED)
```

```python
comp_file.close()
```

```python
zip_obj = zipfile.ZipFile('comp_file.zip','r')
```

```python
zip_obj.extractall('extracted_content')
```
