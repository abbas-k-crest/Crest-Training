# Timing Record

```python
def func_one(n):
    return [str(num) for num in range(n)]
```

```python
func_one(10)
```

```python
def optim(n):
    return list(map(str, range(10)))
```

```python
optim(10)
```

```python
import time
```

```python
# Current Time
start_time = time.time()

#  Run Code
result = func_one(1000000)
end_time = time.time()

# After Code Executed
print("Total Time: ", end_time-start_time)
```

```python
# Current Time
start_time = time.time()

#  Run Code
result = optim(1000000)
end_time = time.time()

# After Code Executed
print("Total Time: ", end_time-start_time)
```

### timeit module

```python
import timeit
```

```python
stmt = """
func_one(100)
"""
```

```python
setup = """
def func_one(n):
    return [str(num) for num in range(n)]
"""
```

```python
timeit.timeit(stmt, setup, number=100000)
```
