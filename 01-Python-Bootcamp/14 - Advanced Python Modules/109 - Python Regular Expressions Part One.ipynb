# REGEX 

```python
import re
```

```python
text = "The agents phone number phone is 408-55-666 phone"
```

```python
pattern = 'phone'
```

```python
re.search(pattern, text)
```

```python
match = re.search(pattern, text)
```

```python
match.start()
```

```python
match.end()
```

```python
match
```

```python
matches = re.findall('phone', text)
```

```python
matches
```

```python
len(matches)
```

```python
for match in re.finditer(pattern, text):
    print(match.group())
```
