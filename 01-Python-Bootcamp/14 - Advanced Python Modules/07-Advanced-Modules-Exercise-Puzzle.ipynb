```python
base_path = 'C:\\Crest Taining\\01-Python-Bootcamp\\14 - Advanced Python Modules\\puzzle_unzip\\extracted_content'
```

```python
# Build regex query
import re

query = r'\d{3}-\d{3}-\d{4}'
```

```python
valid_dir_names = ['One', 'Two', 'Three', 'Four', 'Five']
```

```python
found = False

for root, dir_names, filenames in os.walk(base_path):

    if found:
        break

    # Process only directories we care about
    if os.path.basename(root) not in valid_dir_names:
        print(os.path.basename(root))
        continue

    print("Inside Directory:", root)
    print("Total files:", len(filenames))
    for filename in filenames:

        full_path = os.path.join(root, filename)
        print("Opening file:", filename)

        with open(full_path, "r", errors="ignore") as f:
            text = f.read()
            matches = re.findall(query, text)
        
            if matches:
                print("Found Phone Number:", matches)
                found = True
                break
```
