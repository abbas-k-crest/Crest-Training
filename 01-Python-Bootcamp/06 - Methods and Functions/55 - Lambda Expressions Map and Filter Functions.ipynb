### MAP Function

```python
map
```

```python
def square(num):
    return num**2
```

```python
nums = [1,2,3,4,5]
```

```python
map(square, nums)
```

```python
for item in  map(square, nums):
    print(item)
```

```python
list(map(square, nums))
```

### Filter

```python
def check_even(num):
    return num % 2 == 0
```

```python
filter(check_even, nums)
```

```python
list(filter(check_even, nums))
```

### Lambda Function

```python
lambda num: num**2
```

```python
square = lambda num: num**2
```

```python
square(3)
```

```python
list(map(lambda num: num**2, nums))
```

```python
list(filter(lambda num: num%2==0, nums))
```
