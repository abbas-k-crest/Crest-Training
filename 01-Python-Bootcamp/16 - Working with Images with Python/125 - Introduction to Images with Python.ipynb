### Color Transparency

```python
red = Image.open('red_color.jpg')
```

```python
red
```

```python
blue = Image.open('blue_color.png')
```

```python
blue
```

```python
red.putalpha(128)
red = red.convert("RGBA")
```

```python
red
```

```python
blue.putalpha(128)
blue = blue.convert("RGBA")
```

```python
blue
```

```python
blue.paste(im=red, box=((0,0)), mask=red)
```

```python
blue
```

```python
blue.save("purple.png")
```
