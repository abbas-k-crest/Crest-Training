### try and except block

```python
try:
    print("Hello")
except:
    print("Error occured")
```

```python
try:
    a = 1/0
    print("Hello")
except:
    print("Error occured")
```

```python
try:
    with open("xyz.txt", "w") as file:
        file.readlines()
    print("File read successsully")
except FileNotFoundError as e:
    print(str(e))
```

```python
try:
    with open("xyz.txt", "w") as file:
        file.readlines()
    print("File read successsully")
except FileNotFoundError as e:
    print(str(e))
except FileExistsError as e:
    print("Error caught here")
    print(str(e))
except Exception as e:
    print("Error in Exception")
    print(str(e))
```

```python
try:
    f = open('testfile','w')
    f.write('Test write this')
except IOError:
    # This will only check for an IOError exception and then execute this print statement
    print("Error: Could not find file or read data")
else:
    print("Content written successfully")
    f.close()
```

```python
try:
    with open("testfile", "a") as f:
        f.write("\nHello there")
except IOError as e:
    print(f"IO Error: {str(e)}")
```

```python
try:
    with open("testfile", "r") as f:
        text = f.read()
    print(text)
except IOError as e:
    print(str(e))
except Exception as e:
    print(str(e))
```

```python
try:
    with open("testfile", "r") as f:
        lines = f.readlines()
    print(lines)
    print("successfull")
except Exception as e:
    print(str(e))
```

### try except else

```python
try:
    with open("testfile") as f:
        text = f.read()
except:
    print("Unable to read")
else:
    print("Readed successfully")
```

```python
try:
    with open("testfile") as f:
        f.write("\n I'm writing in the read mode")

except IOError as e:
    print(e)
else:
    print("Written Successfully")
```

### try except finally

```python
try:
    f = open("testfile", "w")
    f.write("\n Here I'm writing to test finally block")
    print("Written Successfully")
except:
    print("Failed")

finally:
    f.close()
    print("Will Run always")
```

```python
try:
    f = open("testfile", "r")
    f.write("\n Here I'm writing to test finally block")
    print("Written Successfully")

except:
    print("Failed")
finally:
    f.close()
    print("Always run even fails")
        
```

```python
try:
    f = open("testfile", "w")
    f.write("Test write statement")
    f.close()
finally:
    print("Always execute finally code blocks")
```

```python
def ask_int():

    try:
        val = int(input("Enter integer here"))
        print("You Entered:", val)
    except:
        print("Looks like you didn't enter integer")
        val = input("Enter integer here")
    finally:
        print("I accept you input:", val)
```
