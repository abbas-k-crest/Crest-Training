## Operators

range()

```python
for num in range(10):
    print(num)
```

```python
for num in range(1, 11):
    print(num)
```

```python
for num in range(1, 11, 2):
    print(num)
```

enumerate

```python
mlist = [11,22,33,44,55,66,77,88,99,111]
```

```python
for idx, num in enumerate(mlist):
    print(f"index: {idx}, Value: {num}")
```

zip 

```python
list1 = [1,2,3,4]

list2 = ['one', 'two', 'three', 'four']
```

```python
for item in zip(list1, list2):
    print(item)
```
