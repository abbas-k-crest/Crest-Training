22. Sum of ASCII for Uppercase Letters

Write a Python program to compute the sum of the ASCII values of the upper-case characters in a given string.
Input:
PytHon ExerciSEs
Output:
373
Input:
JavaScript
Output:
157

```python
def sum_uppercase_ascii(s):
    total = 0
    for ch in s:
        if ch.isupper():
            total += ord(ch)
    return total
```

```python
print(f"SUM", sum_uppercase_ascii("Hello World In"))
```
