# Ex - 1: Merge two lists into a dictionary, using one as keys and the other as values.

```python
def merge_list(l1, l2):

    return dict(zip(l1, l2))
```

```python
keys = ['a', 'b', 'c']
values = [1, 2, 3]
```

```python
merge_list(keys, values)
```

# Ex - 2: Replace all even numbers in a list with the next odd number.

```python
def replace_even(lst):

    return [val+1 if val%2==0 else val for val in lst]
```

```python
lst = [1,2,3,4,5,6,7,8,8,10]
```

```python
replace_even(lst)
```
