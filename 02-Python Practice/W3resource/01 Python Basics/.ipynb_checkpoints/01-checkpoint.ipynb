# 22. Sum of ASCII for Uppercase Letters

### Write a Python program to compute the sum of the ASCII values of the upper-case characters in a given string.
#### Input:Python Exercise
Output:
373
Input:
JavaScript
Output:
157

```python
def sum_uppercase_ascii(s):
    total = 0
    for ch in s:
        if ch.isupper():
            total += ord(ch)
    return total
```

```python
print(f"SUM", sum_uppercase_ascii("Hello World In"))
```

# Sum of First i Equals iÂ¶
Write a Python program to check a given list of integers where the sum of the first i integers is i. Input: [0, 1, 2, 3, 4, 5] Output: False Input: [1, 1, 1, 1, 1, 1] Output: True Input: [2, 2, 2, 2, 2] Output: False

```python
def check_sum(lst):
    total = 0
    for idx, val in enumerate(lst,start=1):
        total += val
        if idx != total:
            print(idx, total)
            return False
            
    return True
```

```python
list1 = [0, 1, 2, 3, 4, 5]
```

```python
print("For list1:", check_sum(list1))
```

```python
list2 = [1, 1, 1, 1, 1, 1]
```

```python
print("For list2:", check_sum(list2))
```

```python
list3 = [2, 2, 2, 2, 2]
```

```python
print("For list3:", check_sum(list3))
```
