## Create a Decorator to Measure Function Execution Time

```python
import time
```

```python
def measure_time(func):

    def wrapper(*args, **kwargs):
        # get time before function execution
        start_time = time.time()

        # run function
        result = func(*args, **kwargs)

        # get time after function execution
        end_time = time.time()

        total_time = end_time - start_time
        print(f"Total time by {func.__name__}: {total_time:.2f} seconds")

        return result

    return wrapper
```

```python
@measure_time
def slow_func():
    time.sleep(2)
    print("Finished Execution")
```

```python
slow_func()
```
