## Unit Test for Sorted List in Ascending Order

```python
import unittest
```

```python
def is_sorted_list(lst):

    for i in range(1, len(lst)):
        if lst[i-1] > lst[i]:
            return False
    return True
```

```python
class TestSortedList(unittest.TestCase):

    def test_sorted(self):
        self.assertTrue(is_sorted_list([1,2,3,4,5,6]))

    def test_unsorted(self):
        self.assertFalse(is_sorted_list([1,2,3,4,2,1]))

    def test_empty(self):
        self.assertTrue(is_sorted_list([]))

    def test_single_element(self):
        self.assertTrue(is_sorted_list([42]))

    def test_duplicates(self):
        self.assertTrue(is_sorted_list([1, 2, 2, 3, 4]))
```

```python
unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
