## Convert timezone from local to utc, utc to local or specified zones

```python
import arrow
```

```python
utc = arrow.utcnow()
print("utc:")
print(utc)
```

```python
print("\nutc to local:")
print(utc.to('local'))
```

```python
print("\nlocal to utc:")
print(utc.to('local').to('utc'))
```

```python
print("\nutc to specific location:")
print(utc.to('US/Pacific'))
```
