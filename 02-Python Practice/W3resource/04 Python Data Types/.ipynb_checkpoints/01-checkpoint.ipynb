## Collections: Find Most Common Elements and Counts in a Text

```python
from collections import Counter
```

```python
def most_common(s):
    return Counter(s).most_common()
```

```python
s = "lkseropewdssafsdfafkpwe"
```

```python
most_common(s)
```

```python
most_common(s)[0]
```

## JSON: Convert Python Dictionary (Sorted by Key) to JSON Data

```python
import json
```

```python
def dict_to_json(d):
    return json.dumps(d, sort_keys=True)
```

```python
data = {
    "Apple": "Red",
    "Banana": "Yellow",
    "Orange": "Orange",
    "Watermelon": "Green"
}
```

```python
dict_to_json(data)
```

## Array: Count occurrences of a specified element in an array

```python
def count_occurence(arr, k):
    return arr.count(k)
```

```python
arr = [1,2,3,4,3,2,5,4,6,4,3,2,4,5,3,3,2,1,2,3,4,4,7,2,2,5]
```

```python
count_occurence(arr, 1)
```

```python
count_occurence(arr, 2)
```

```python
count_occurence(arr, 4)
```

```python
count_occurence(arr, 3)
```

## Enum: Display Enum member names ordered by their values

```python
import enum
```

```python
class CountryCode(enum.IntEnum):
    India = 91
    UAE = 971
    Bahrain = 973
    Albania = 355
```

```python
print("Country Codes:")
print('\n'.join(str(c) for c in sorted(CountryCode)))
```
