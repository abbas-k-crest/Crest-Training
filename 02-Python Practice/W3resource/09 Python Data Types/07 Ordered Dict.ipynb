## OrderedDict from Random Data

```python
import random
from collections import OrderedDict
```

```python
def random_ascii():
    return chr(random.randint(65, 90))
```

```python
ordered_dict = OrderedDict()
```

```python
for i in range(10):
    key = random_ascii()
    value = random.randint(1, 50)
    ordered_dict[key] = value
```
