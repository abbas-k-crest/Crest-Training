## Get a list of dates between two dates

```python
from datetime import timedelta, date
```

```python
def get_date_range(date1, date2):

    for n in range(int((date2-date1).days)+1):
        yield date1 + timedelta(n)
```

```python
start = date(2025, 10, 20)
end = date(2025, 11, 30)
```

```python
for dt in get_date_range(start, end):
    print(dt)
```

```python
from datetime import datetime, timedelta

start_date = datetime.strptime("2025-08-10", "%Y-%m-%d")
end_date = datetime.strptime("2025-08-15", "%Y-%m-%d")

date_list = [(start_date + timedelta(days=i)).date() for i in range((end_date - start_date).days + 1)]

print(date_list)
```
