# Matrix Multiplication via List Comprehensions

```python
A = [
    [1, 2],
    [3, 4]
]

B = [
    [5, 6],
    [7, 8]
]
```

```python
def matrix_mul(a, b):
    rows_a = len(a)
    cols_a = len(a[0])
    cols_b = len(b[0])
    
    result = [
        [sum(a[i][k] * b[k][j]  for k in range(cols_a)) for j in range(cols_b)]
        for i in range(rows_a)
    ]
    
    return result
```

```python
matrix_mul(A, B)
```

```python
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

B = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

result = matrix_mul(A, B)
print(result)
```
