# Ex - 4: Student Management (CRUD)

```python
class Student:

    def __init__(self, name, enroll_num, standard):
        self.name=name
        self.enroll_num=enroll_num
        self.standard=standard

    def update(self, name=None, enroll_num=None, standard=None):
        if not name and not enroll_num and not standard:
            return f"Nothing to update"

        if name:
            prev, self.name = self.name, name
            return f"Name changed from {prev} to {self.name}"
        if enroll_num:
            prev, self.enroll_num = self.enroll_num, enroll_num
            return f"Enrollment No. changed from {prev} to {self.enroll_num}"
        if standard:
            prev, self.standard = self.standard, standard
            return f"Standard changed from {prev} to {self.standard}"
```

```python
std1 = Student("Abbas Kothari", 122001, 'I')
std2 = Student("Harsh Mistry", 122002, 'II')
std3 = Student("Maiterya Saparia", 122003, 'III')
```

```python
std1.update(standard='III')
```
