# Ex - 1: Search an Element in a Circular Linked List

```python
class Node:

    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
head = Node(10)
one = Node(11)
two = Node(12)
three = Node(13)
```

```python
head.next = one
one.next = two
two.next = three
three.next = head
```

```python
def search_element(num):
    if head.data == num:
        return True
    node = head.next
    
    while True:
        if node.data == num:
            return True
        if node.next == head:
            return False

        node = node.next
```

```python
search_element(10)
```

```python
search_element(12)
```

```python
search_element(19)
```
