# Ex -1 Counting Sort

```python
def count_sort(s):
    count = [0] * 256
    
    for char in s:
        count[ord(char)] += 1

    sorted_str = ""
    for i in range(256):
        sorted_str += chr(i) * count[i]

    return sorted_str
```

```python
s = "stringggg"
print(count_sort(s))
```

```python
s = "vwyvyewbuwhvbuwbvuwbvwubv"

print(count_sort(s))
```

# Ex -2: Brick Sort

```python
def oddEvenSort(arr):

    isSorted = False

    while not isSorted:

        isSorted = True

        for i in range(0, len(arr)-1, 2):
            if arr[i] > arr[i+1]:
                arr[i], arr[i+1] = arr[i+1], arr[i]
                isSorted = False

        for i in range(1, len(arr)-1, 2):
            if arr[i] > arr[i+1]:
                arr[i], arr[i+1] = arr[i+1], arr[i]
                isSorted = False
```

```python
arr = [3, 4, 2, 10, 9] 
oddEvenSort(arr); 
for i in range(len(arr)): 
    print(arr[i], end= " ")
```

# Ex - 3: Gnome Sort

```python
def gnomeSort(arr):

    n = len(arr)
    idx = 0
    
    while idx < n:
        print(idx)
        if idx == 0:
            idx += 1
            continue
        if arr[idx] >= arr[idx-1]:
            idx +=1

        else:
            arr[idx-1], arr[idx] = arr[idx], arr[idx-1]
            idx -=1
        
```

```python
arr = [3, 4, 2, 10, 9] 
```

```python
gnomeSort(arr)
```

# Ex - 4: Comb Sort

```python

arr = [3, 4, 2, 10, 9] 
n = len(arr)
gap = n
shrink = 1.3
swapped = True

while gap > 1 or swapped:
    gap = max(1, int(gap / shrink))
    swapped = False

    for i in range(n - gap):
        if arr[i] > arr[i + gap]:
            arr[i], arr[i + gap] = arr[i + gap], arr[i]
            swapped = True
```

# Ex - 5: Shell Sort

```python
def shellSort(arr):
    n = len(arr)
    gap = n//2

    while gap > 0:
        for i in range(gap,n):
            temp = arr[i]
            j = i
            while j >= gap and arr[j-gap] >temp:
                arr[j] = arr[j-gap]
                j -= gap

            arr[j] = temp
        gap //= 2
```

```python
arr = [ 12, 34, 54, 2, 3]
n = len(arr)
shellSort(arr)
for i in range(n):
    print(arr[i])
```
