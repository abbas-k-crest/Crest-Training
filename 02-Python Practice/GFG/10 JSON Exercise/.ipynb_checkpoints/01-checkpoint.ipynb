# Convert class object JSON in Python

```python
import json
```

```python
class Person():

    def __init__(self, name, age):
        self.name=name
        self.age=age
```

```python
p1 = Person("Abbas", 22)
p2 = Person("Harsh", 22)
p3 = Person("Pratik", 22)
```

```python
di = json.dumps(p1.__dict__)
```
